```python
import gzip
import json
import pandas as pd

def parse_json_gz(path):
    data = []
    with gzip.open(path, 'rt', encoding='utf-8') as f:  # Open gzipped file in text mode
        for line in f:
            data.append(json.loads(line.strip()))  # Load each line as a separate JSON object
    return data

# Specify your file path
file_path = "meta-Texas.json.gz"

# Parse JSON data
data = parse_json_gz(file_path)

# Convert to DataFrame
df = pd.DataFrame(data)

# Display first few rows
print(df.head())
```

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary resources from nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize Lemmatizer and stopwords list
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    """
    Function to preprocess text by performing:
    1. Tokenization
    2. Lowercasing
    3. Removing stopwords
    4. Lemmatization
    """
    # Tokenize the text
    tokens = word_tokenize(text.lower())  # Convert to lower case before tokenization
    
    # Remove punctuation
    tokens = [word for word in tokens if word.isalpha()]
    
    # Remove stop words
    tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatization (reduce words to their base form)
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Return the cleaned and lemmatized text as a string
    return " ".join(lemmatized_tokens)
```

```python
data_cleaned_no_duplicates['cleaned_description'] = data_cleaned_no_duplicates['description'].apply(preprocess_text)
data_cleaned_no_duplicates['cleaned_review'] = data_cleaned_no_duplicates['text'].apply(preprocess_text)
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')
```

```python
# Initialize VADER SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Function to get sentiment
def get_sentiment(text):
    score = sia.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply the sentiment analysis to the 'text' column
data_cleaned_no_duplicates['review_sentiment'] = data_cleaned_no_duplicates['text'].apply(get_sentiment)

# View the results
print(data_cleaned_no_duplicates[['text', 'review_sentiment']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Convert reviews to TF-IDF
tfidf_matrix = vectorizer.fit_transform(data_cleaned_no_duplicates['text'])

# You can view the feature names (terms)
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Function to perform search based on user query
def smart_search(query, tfidf_matrix, data, top_n=5):
    query_vec = vectorizer.transform([query])  # Convert query to TF-IDF vector
    similarities = cosine_similarity(query_vec, tfidf_matrix)  # Compute similarity between query and reviews
    
    # Get top N matching reviews
    top_indices = similarities[0].argsort()[-top_n:][::-1]  # Sort to get highest similarity
    
    # Return the top N business names or any other details
    return data.iloc[top_indices]

# Example search query
query = "best cafés for studying"
top_results = smart_search(query, tfidf_matrix, data_cleaned_no_duplicates)
print(top_results[['name', 'text']])
```
